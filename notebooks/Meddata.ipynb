# Team 6: Predicting Response to Neoadjuvant Chemotherapy to Improve Treatment Decisions
Diana Dadkhah Tirani, Gurveer Madurai, Derrick Jaskiel, Rachel Chiu, Victoria Farkas, Nandini Krishnan

We engineered an end-to-end multimodal deep learning pipeline to predict breast cancer patient responses to Neoadjuvant Chemotherapy (NAC). By extracting spatial features like nuclear clumping from tiled Whole Slide Images (WSIs) and fusing them with traditional clinical data, we trained a custom Keras neural network on the enriched dataset. This multimodal approach successfully outperformed all baseline and tree-based models, proving that integrating digital pathology with clinical history significantly improves predictive accuracy.

### Biopsy Image Feature Extraction

WSI (Whole Slide Images) are too expensive computationally to process as a whole, so a sliding window strategy was used to tile and sample non-background patches of the stained tissue.

<img src="../assets/patch_samples.png" alt="rpatchesc" width="400"/>

Features of interest (edges and nuclear clumps) were identified and extracted into NumPy arrays from the tissue patches, using OpenCV.

- Img_Nuclear_Density (Proliferation)
- Img_Edge_Complexity (Invasiveness)
- Img_Clump_Count_Per_Patch (Tissue Disorganization/Differentiation)

### DL Model Metrics

Our model achieved 78% overall accuracy and demonstrated reliability when predicting a Complete Response (92% precision). It operates with a slightly conservative clinical baseline—excelling at catching almost all Partial Responses (93% recall) while occasionally underestimating patients who actually achieved complete remission, which is the ideal balance for a clinical model.

| Class / Metric | Precision | Recall | F1-Score | Support |
| :--- | :--- | :--- | :--- | :--- |
| **Partial Response (0)** | 0.70 | 0.93 | 0.80 | 15 |
| **Complete Response (1)** | 0.92 | 0.65 | 0.76 | 17 |
| **Accuracy** | - | - | **0.78** | 32 |
| **Macro Avg** | 0.81 | 0.79 | 0.78 | 32 |
| **Weighted Avg** | 0.82 | 0.78 | 0.78 | 32 |
